# 결측치 처리 (예시)

## 자주 쓰는 것은 함수로 만들어 둬도 좋음

```python
def show_nas(df):
    # 결측치 확인하고 개수가 많은 순서로 정렬
    missing_values = df.isnull().sum()
    missing_values.sort_values(ascending=False, inplace=True)

    # 결측치가 있는 열을 반환
    return missing_values[missing_values > 0]
```

```python
show_nas(sp500_info_df)
```

```python
import numpy as np

# payoutRatio가 0인 행을 찾고 해당 행의 dividendYield 또는 fiveYearAvgDividendYield이 Null인 경우 0으로 채웁니다.
sp500_info_df.loc[(sp500_info_df['payoutRatio'] == 0) & (sp500_info_df['dividendYield'].isnull()), 'dividendYield'] = 0
sp500_info_df.loc[(sp500_info_df['payoutRatio'] == 0) & (sp500_info_df['fiveYearAvgDividendYield'].isnull()), 'fiveYearAvgDividendYield'] = 0
```

```python
show_nas(sp500_info_df)
```

```python
# trailingPE가 Null인 행을 찾고 해당 행을 0으로 채웁니다.
sp500_info_df.loc[(sp500_info_df['trailingPE'].isnull()), 'trailingPE'] = 0
```

```python
show_nas(sp500_info_df)
```

```python
#이거는 payoutRatio가 yfinance에 정말 등록 안된 경우도 있는 것 같아서 매뉴얼로 채우기로 함.
sp500_info_df[sp500_info_df['payoutRatio'].isna()][['shortName','dividendYield']]
```
